```python
# python
# BASE から，各国のコンテンツ数をとる
import requests
from bs4 import BeautifulSoup
import datetime
import locale
import re

country = ["jp","us","uk","ca","fr","de","it"]

time = datetime.datetime.now()
print(str(time) + '\n' + '[')

for item in country:
   # r = requests.get('https://www.base-search.net/Search/Results?q=country:jp')
   r = requests.get('https://www.base-search.net/Search/Results?q=country:'+item) 
   data = BeautifulSoup(r.text, 'html.parser')
   # print(data.select("form.ID_f_dctypenorm > option")[0])

   elems = data.find_all("form", id="ID_f_dctypenorm")
   # print(elems ">option" [0])
   elems = elems.find_all("option")
   for e in elems:
      te = e.getText()
      te = te.replace(',','')
      se = te.replace('(', "["+item+",")
      se = se.replace('\n', ']\n')
      se = se.replace(') ', ',')
      itemse = item + se + "]"
      itemse = re.sub('$','[',itemse)
      print(itemse)
```
